```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import sklearn.datasets
```

```python
breast_cancer_data = pd.read_csv('BreastCancerDataset.csv')
```

```python
BreastCancer = sklearn.datasets.load_breast_cancer()
```

```python
print(BreastCancer)
```

```python
data_frame = pd.DataFrame(BreastCancer.data, columns = BreastCancer.feature_names)
```

```python
data_frame.head()
```

```python
data_frame['label']= BreastCancer.target
```

```python
#  1 --> Benign
#  0 --> Malignant

data_frame.groupby('label').mean()
```

```python
#Separating the data into features and targets
X = data_frame.drop(columns = 'label', axis=1)
y = data_frame['label']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state =2)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.fit_transform(X_test)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#Setting up the layers of Neural Networks
model = keras.Sequential([
                           keras.layers.Flatten(input_shape=(30,)),        #Input layer
                           keras.layers.Dense(20, activation = 'relu') ,     #Hidden Layer
                           keras.layers.Dense(2, activation = 'sigmoid')      #OutputLayer
                                    ])
```

```python
#Compiling the neural networks

model.compile(optimizer='adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
#Training the NeuralNetwork
history = model.fit(X_train_std, y_train, validation_split = 0.1,  epochs = 10 )
```

```python
#Visualizing Accuracy and Loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
#Visualizing Accuracy and Loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
#Accuracy of the model on test data
loss, accuracy = model.evaluate(X_test_std, y_test)
print(accuracy)
print(loss)
```
